```python
import pandas
import matplotlib.pyplot as plt
import seaborn
```

```python
# load data using pandas
data=pandas.read_csv('C:/Users/Narjiss/Documents/uber-data.txt')
```

```python
# display the first 5 rows of the dataset
data.head()
```

```python
# select a subset of the data 
# the Lat column
data['Lat']
```

```python
# Convert the "Date/Time" column to datetime
data['Date/Time']=data['Date/Time'].map(pandas.to_datetime)
```

```python
data.head()
```

```python
# define a function that takes a datetime as an argument and returns the corresponding day of month
def get_dom(dt):
    return dt.day

# add a new column named "dom" that contains the day of month
data['dom']=data['Date/Time'].map(get_dom)
```

```python
data.head()
```

```python
# define a function that takes a datetime as an argument and returns the corresponding weekday
def get_weekday(dt):
    return dt.weekday()

# add a new column named "weekday" that contains the day of the week
data['weekday']=data['Date/Time'].map(get_weekday)
```

```python
data.head()
```

```python
# define a function that takes a datetime as an argument and returns the corresponding hour
def get_hour(dt):
    return dt.hour

# add a new column named "hour" that contains the hour
data['hour']=data['Date/Time'].map(get_hour)
data.head()
```

# Analysis

## Analysis of DoM

```python
# plot a histogram of the "dom" column

data.hist('dom', bins=30,width=0.8,range=(0.5,30.5))
plt.xlabel("Day of Month")
plt.ylabel("Frequency")
plt.title("Frequency by DoM - uber - Apr 2014")
```

```python
'''for k, rows in data.groupby('dom'):
    print((k,len(rows)))
'''
def count_rows(rows):
    return len(rows)
by_date=data.groupby('dom').apply(count_rows)
by_date
```

```python
plt.plot(by_date)
```

```python
data.hist(column='dom', bins=30,width=0.8,range=(0.5,30.5))
plt.xlabel("Day of Month")
plt.ylabel("Frequency")
plt.title("Frequency by DoM - uber - Apr 2014")
```

```python
plt.bar(range(1,31),by_date)
```

```python
by_date_sorted=by_date.sort_values()
by_date_sorted
```

```python
plt.bar(range(1,31),by_date_sorted)
plt.xticks(range(1,31),by_date_sorted.index)
plt.xlabel("Day of Month")
plt.ylabel("Frequency")
plt.title("Frequency by DoM - uber")
;
```
